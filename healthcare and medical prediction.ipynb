```python
bp=['Normal','Low','High','Normal','High','High','High']
pulse=['Normal','High','High','Normal','High','High','High']
HB=['Normal','Low','Normal','Normal','Normal','Normal','Low']
TLC=['Normal','Low','Normal','Normal','Normal','Normal','Normal']
Platelets=['Normal','Low','Normal','Normal','Normal','Normal','Normal']
sugar_fasting=['Normal','High','High','High','High','High','High']
sugar_pp=['Normal','Low','High','High','Normal','High','High']
HBA1C=['Normal','Normal','High','High','Normal','High','High']
Cholestrol=['Normal','Normal','High','Normal','High','High','High']
Triglycerides=['Normal','Normal','High','High','High','High','High']
ECG=['Normal','Normal','Abnormal','Normal','Abnormal','Abnormal','Abnormal']
heart_disease=['No','No','Yes','No','No','Yes','Yes']
diabeites=['No','No','No','Yes','No','Yes','Yes']

from sklearn import preprocessing
#creating labelEncoder
le = preprocessing.LabelEncoder()
# Converting string labels into numbers.
bp_encoded=le.fit_transform(bp)
print ("bp_encoded",bp_encoded)
pulse_encoded=le.fit_transform(pulse)
print ("pulse_encoded:",pulse_encoded)
HB_encoded=le.fit_transform(HB)
print ("HB_encoded",HB_encoded)
TLC_encoded=le.fit_transform(TLC)
print ("TLC_encoded",TLC_encoded)
platelets_encoded=le.fit_transform(Platelets)
print ("platelets_encoded",platelets_encoded)
sugar_fasting_encoded=le.fit_transform(sugar_fasting)
print ("sugar_fasting_encoded",sugar_fasting_encoded)
sugar_pp_encoded=le.fit_transform(sugar_pp)
print ("sugar_pp_encoded",sugar_pp_encoded)
HBA1C_encoded=le.fit_transform(HBA1C)
print ("HBA1C_encoded",HBA1C_encoded)
Cholestrol_encoded=le.fit_transform(Cholestrol)
print ("Cholestrol_encoded",Cholestrol_encoded)

Triglycerides_encoded=le.fit_transform(Triglycerides)
print ("Triglycerides_encoded",Triglycerides_encoded)
ECG_encoded=le.fit_transform(ECG)
print ("ECG_encoded",ECG_encoded)

heart_disease_encoded=le.fit_transform(heart_disease)
print ("heart_disease_encoded",heart_disease_encoded)
diabeites_encoded=le.fit_transform(diabeites)
print ("diabeites_encoded",diabeites_encoded)

#put them in one list using zip
tests=list(zip(bp_encoded,pulse_encoded,HB_encoded,TLC_encoded,platelets_encoded,sugar_fasting_encoded,sugar_pp_encoded,HBA1C_encoded,Cholestrol_encoded,Triglycerides_encoded,ECG_encoded))
print (tests)


bp=input("enter value ofbp:")
if ((bp>='100/60')and(bp<='138/88')):
    bp=2
elif(bp<'100/60'):
    bp=1
else:
    bp=0
pulse=eval(input("enter value of pulse"))
if(60<=pulse<=88):
    pulse=1
elif(pulse>88):
    pulse=0
else:
    pulse=2
HB=eval(input("enter value of hb"))
if(12<=HB<=14):
    HB=1
elif(HB<12):
    HB=0
else:
    HB=2

TLC=eval(input("enter value of tlc"))
if(4000<=TLC<=6000):
    TLC=1
elif(TLC<4000):
    TLC=0
else:
    TLC=2
platelet=eval(input("enter value of platelets:"))
if(100000<=platelet<=500000):
    platelet=1
elif(platelet<100000):
    platelet=0
else:
    platelet=2
sugar_fasting=eval(input("enter value of sugar fasting"))
if(60<sugar_fasting<90):
    sugar_fasting=1
elif(sugar_fasting>90):
    sugar_fasting=0
else:
    sugar_fasting=2
sugar_pp=eval(input("enter value of sugar pp"))
if(100<=sugar_pp<=140):
    sugar_pp=2
elif(sugar_pp<100):
    sugar_pp=1
else:
    sugar_pp=0
HBA1C=eval(input("enter value of hba1c"))
if(0<=HBA1C<=6):
    HBA1C=1
if(HBA1C>=6):
    HBA1C=0
else:
    HBA1C=2
Cholestrol=eval(input("enter value of Cholestrol"))
if(150<=Cholestrol<=200):
    Cholestrol=1
elif(Cholestrol>200):
    Cholestrol=0
else:
    Cholestrol=2

Triglycerides=eval(input("enter value of triglycerides"))
if(150<=Triglycerides<=199):
    Triglycerides=1
elif(Triglycerides>199):
    Triglycerides=0
else:
    Triglycerides=2

ECG=input("enter value of ECG")
if(ECG=='Normal'):
    ECG=1
elif(ECG=='Abnormal'):
    ECG=0
from sklearn.naive_bayes import GaussianNB
def heart():
    le = preprocessing.LabelEncoder()
    heart_disease_encoded=le.fit_transform(heart_disease)
    print ("heart_disease_encoded",heart_disease_encoded)
    model = GaussianNB()
    model.fit(tests,heart_disease_encoded)
    #prediction;
    predicted= model.predict([[bp,pulse,HB,TLC,platelet,sugar_fasting,sugar_pp,HBA1C,Cholestrol,Triglycerides,ECG]])
    print ("heart disease Predicted Value:", predicted)
heart()
```
